```python
#importing the panda library & creating the variable using the panda in csv file
import pandas as pd
dataset=pd.read_csv("salarydata.csv")
```

```python
# Running the dataset
dataset
```

```python
# getting the input value in the dataset  using independent variable
independent=dataset[["Experience"]]
independent
```

```python
# getting the output value in the dataset  using dependent variable
dependent=dataset[["Salary"]]
dependent
```

```python
#To split the train & test value using the sklearn module using the modelselection 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#to create the model creation using the sklearn module using linear model option or folder for getting the weight & bias value
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
#to getting the weight using the regressor class in coef function
weight=regressor.coef_
weight
```

```python
#to getting the bias using the regressor class in intersept function
bias=regressor.intercept_
bias
```

```python
# to predict the output test value using the predict function
Y_pred=regressor.predict(X_test)
```

```python
#to validate the actual output & prdedicted output using the metrics module(near by 1 is good & nearby 0 is poor prdeiction)
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
r_score
```

```python
# import the pickle option to save the maodel & getting the result based on the input
import pickle
filename="finalized_model_Linear.sav"
pickle.dump(regressor,open(filename,'wb'))
loaded_model=pickle.load(open("finalized_model_Linear.sav",'rb'))
result=loaded_model.predict([[int(input("Enter the years of Experience:"))]])
result
```
